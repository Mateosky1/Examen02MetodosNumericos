# Splines cúbicos

```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.show()
```

# Tarea
* Grafique los splines cúbicos tal que cumplan con la pendiente $m=10$ deseada en el punto $(x_1,y_1)$.

```python
import matplotlib.pyplot as plt
import numpy as np

def Spline(x: float, x0: float, pars: dict) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3

xs = [-1, 0, 1]
ys = [1, 5, 3]

s = [
    {"a": 1, "b": 1, "c": 0, "d": 3},        
    {"a": 5, "b": 10, "c": -18, "d": 6},     
]

plt.figure(figsize=(8, 6))

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 50)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="blue", linestyle="--", linewidth=2)

plt.scatter(xs, ys, color="tab:blue", s=50, zorder=5)
plt.text(0.45, 4.8, r"$m_{deseada} = 10$", fontsize=11)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.grid(False)
plt.xlim(-1.1, 1.1)
plt.ylim(0.8, 6.8)
plt.show()
```
